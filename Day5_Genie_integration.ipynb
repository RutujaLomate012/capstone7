```python
%sql
CREATE OR REPLACE VIEW knowledgehub_lakehouse.gold.vw_doc_freshness AS
SELECT
  doc_id,
  department_key,
  freshness_days,
  CASE
    WHEN freshness_days > 365 THEN 'STALE'
    ELSE 'FRESH'
  END AS freshness_status
FROM knowledgehub_lakehouse.gold.fact_doc_versions
WHERE current_version_flag = true;
```

```python
%sql
CREATE OR REPLACE VIEW knowledgehub_lakehouse.gold.vw_compliance_risk AS
SELECT
  c.doc_id,
  d.department,
  c.current_score
FROM knowledgehub_lakehouse.gold.compliance_risk_summary c
JOIN knowledgehub_lakehouse.silver.docs_clean d
  ON c.doc_id = d.doc_id
WHERE d.current_version_flag = true;
```

```python
%sql
CREATE OR REPLACE VIEW knowledgehub_lakehouse.gold.vw_doc_usage AS
SELECT
  f.doc_id,
  f.event_date,
  f.views,
  f.downloads,
  f.searches,
  d.search_success_rate_7d
FROM knowledgehub_lakehouse.gold.fact_doc_daily_usage f
JOIN knowledgehub_lakehouse.silver.doc_features d
  ON f.doc_id = d.doc_id;
```

```python
%sql
CREATE OR REPLACE VIEW knowledgehub_lakehouse.gold.mv_daily_usage AS
SELECT
  event_date,
  SUM(views) AS total_views,
  SUM(downloads) AS total_downloads
FROM knowledgehub_lakehouse.gold.fact_doc_daily_usage
GROUP BY event_date;
```
